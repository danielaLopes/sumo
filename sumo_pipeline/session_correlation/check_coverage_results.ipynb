```python
import pickle
```

```python
DATA_RESULTS_FOLDER = 'results/data/'
```

```python
results_by_eu_country_file = f'results_by_eu_country_small_OSTest.pickle'
results_by_zone1 = pickle.load(open(DATA_RESULTS_FOLDER+results_by_eu_country_file, 'rb'))
print("results_by_zone full coverage", results_by_zone1['Full coverage'])
```

```python
results_by_eu_country_file = f'results_by_eu_country_small_OSTest.pickle1'
results_by_zone2 = pickle.load(open(DATA_RESULTS_FOLDER+results_by_eu_country_file, 'rb'))
print("results_by_zone full coverage", results_by_zone2['Full coverage'])
```

```python
results_by_eu_country_file = f'results_by_eu_country_small_OSTest.pickle2'
results_by_zone3 = pickle.load(open(DATA_RESULTS_FOLDER+results_by_eu_country_file, 'rb'))
print("results_by_zone full coverage", results_by_zone3['Full coverage'])
```

```python
if results_by_zone1['Full coverage'] == results_by_zone2['Full coverage'] == results_by_zone3['Full coverage']:
    print("All dictionaries are equal")
else:
    print("Dictionaries are not equal")
```

```python
different_keys1 = []

for key in results_by_zone1['Full coverage']:
    if results_by_zone1['Full coverage'][key] != results_by_zone2['Full coverage'][key]:
        different_keys1.append(key)
        print("\n--- results_by_zone1[key]", key, results_by_zone1['Full coverage'][key])
        print("results_by_zone1[key]", key, results_by_zone2['Full coverage'][key])

print("Keys with different values:", different_keys1)
```

```python
different_keys2 = []

for key in results_by_zone2['Full coverage']:
    if results_by_zone2['Full coverage'][key] != results_by_zone3['Full coverage'][key]:
        different_keys2.append(key)
        print("\n--- results_by_zone2[key]", key, results_by_zone2['Full coverage'][key])
        print("results_by_zone3[key]", key, results_by_zone3['Full coverage'][key])

print("Keys with different values:", different_keys2)
```
